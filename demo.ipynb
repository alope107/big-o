```python
%matplotlib inline
from complex_viz.tracker import Tracker
from complex_viz.plotter import compare_growth, plot_growth
```

```python
tracker = Tracker()
```

```python
def single_loop(n):
    total = 0
    for i in range(n):
        tracker.add()
        total += i
    return total

def double_loop(n):
    total = 0
    for i in range(n):
        for j in range(n):
            tracker.add()
            total += i
    return total

def triple_loop(n):
    total = 0
    for i in range(n):
        for j in range(n):
            for k in range(n):
                tracker.add()
                total += i
    return total

```

```python
compare_growth((single_loop, double_loop, triple_loop), list(range(10)), tracker)
```
