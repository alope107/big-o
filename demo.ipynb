```python
%matplotlib inline
from complex_viz.baselines import linear, quadratic, cubic, log, nlogn, factorial
from complex_viz.plotter import compare_growth, plot_growth
```

```python
def single_loop(n, tracker):
    total = 0
    for i in range(n):
        tracker.add()
        total += i
    return total

def double_loop(n, tracker):
    total = 0
    for i in range(n):
        for j in range(n):
            tracker.add()
            total += i
    return total

def triple_loop(n, tracker):
    total = 0
    for i in range(n):
        for j in range(n):
            for k in range(n):
                tracker.add()
                total += i
    return total
```

```python
compare_growth((single_loop, double_loop, triple_loop), range(0, 10))
```

```python
compare_growth((linear, log, nlogn), range(0, 40))
```

```python
plot_growth(cubic, range(10))
```
